A) Ejercicio 1: Suma de imágenes.

```python
#Librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar las dos imágenes
imagen2 = cv2.imread("RXmano.PNG", cv2.IMREAD_GRAYSCALE)
imagen1 = cv2.imread("RXtoracica.PNG", cv2.IMREAD_GRAYSCALE)

# Redimensionar ambas imágenes al mismo tamaño si es necesario
imagen2_res = cv2.resize(imagen2, (imagen1.shape[1], imagen1.shape[0]))

# Aplicar la suma ponderada con pesos α=0.7 y β=0.3
alpha = 0.3  # Peso de la primera imagen
beta = 0.7   # Peso de la segunda imagen
gamma = 10     # Ajuste de brillo

imagen_resultado = cv2.addWeighted(imagen1, alpha, imagen2_res, beta, gamma)

# Cargar varias imágenes en una lista
imagenes = ["RXtoracica.PNG", "RXmano.PNG"]
imagenes_cargadas = [cv2.imread(img, cv2.IMREAD_GRAYSCALE) for img in imagenes]

# Verificar que todas las imágenes tengan el mismo tamaño
alto, ancho = imagenes_cargadas[0].shape
imagenes_cargadas = [cv2.resize(img, (ancho, alto)) for img in imagenes_cargadas]

# Crear una imagen acumuladora en formato flotante
suma = np.zeros((alto, ancho), dtype=np.float32)

# Sumar todas las imágenes
for img in imagenes_cargadas:
    suma += img

# Calcular el promedio dividiendo por la cantidad de imágenes
imagen_promedio = suma / len(imagenes_cargadas)

# Convertir la imagen a formato uint8 para visualización
imagen_promedio = cv2.convertScaleAbs(imagen_promedio)

# Mostrar imágenes originales
for i, img in enumerate(imagenes_cargadas):
    plt.subplot(1, len(imagenes), i+1)  # Crear subgráficos en una fila
    #plt.figure(figsize=(10, 5))
    plt.imshow(img, cmap="gray")  # Mostrar imagen
    plt.title(f"Imagen {i+1}")  # Título de cada imagen
    plt.axis("off")  # Ocultar ejes

plt.show()

#Mostrar imagen promedio
plt.figure(figsize=(5, 5))
#plt.subplot(1, 3, 4)
plt.imshow(imagen_promedio, cmap="gray")
plt.title("Imagen Promedio")
plt.axis("off")
plt.show()

# Mostrar imágenes
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(imagen1, cmap="gray")
plt.title("Imagen Original 1")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(imagen2, cmap="gray")
plt.title("Imagen Original 2")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(imagen_resultado, cmap="gray")
plt.title("Resultado Suma Ponderada")
plt.axis("off")

plt.show()
```

B) Ejercicio 2: Resta de imágenes.

```python
# Cargar las dos imágenes
img1 = cv2.imread("RXtoracica.PNG", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("RXmano.PNG", cv2.IMREAD_GRAYSCALE)

# Asegurar que tengan el mismo tamaño
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Realizar la resta absoluta de imágenes
resta = cv2.absdiff(img2, img1)

# Mostrar imágenes originales y resultado
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title("Imagen 1")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
plt.title("Imagen 2")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(resta, cmap='gray')
plt.title("Imagen Restada (AbsDiff)")
plt.axis("off")

plt.show()
```

C) Ejercicio 3: Multiplicación por un factor.



```python
# Cargar imagen en escala de grises
img = cv2.imread("RXmano2.PNG", cv2.IMREAD_GRAYSCALE)

# Definir el factor de multiplicación
factor1 = 0.5  # Aumenta el brillo en un 50%
factor2 = 1  # Aumenta el brillo en un 50%
factor3 = 1.5  # Aumenta el brillo en un 50%

# Multiplicar la imagen por el factor
img_multiplicada1 = cv2.multiply(img, np.array([factor1], dtype=np.float32))
img_multiplicada2 = cv2.multiply(img, np.array([factor2], dtype=np.float32))
img_multiplicada3 = cv2.multiply(img, np.array([factor3], dtype=np.float32))

# Convertir la imagen a uint8 para visualización correcta
img_multiplicada1 = cv2.convertScaleAbs(img_multiplicada1)
img_multiplicada2 = cv2.convertScaleAbs(img_multiplicada2)
img_multiplicada3 = cv2.convertScaleAbs(img_multiplicada3)

# Mostrar la imagen original y la modificada
plt.figure(figsize=(10, 4))

plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(img_multiplicada1, cmap='gray')
plt.title(f"Imagen x {factor1}")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(img_multiplicada2, cmap='gray')
plt.title(f"Imagen x {factor2}")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(img_multiplicada3, cmap='gray')
plt.title(f"Imagen x {factor3}")
plt.axis("off")


plt.show()
```

D) Ejercicio 4: Multiplicación elemento por elemento

A) Ejercicio 5: Operaciones lógicas AND, OR, XOR y NOT

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar las dos imágenes en escala de grises
img1 = cv2.imread("/content/figmano2.jpeg", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("/content/figmano1.jpg", cv2.IMREAD_GRAYSCALE)

# Redimensionar img2 para que tenga el mismo tamaño que img1
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Aplicar operaciones lógicas directamente a las imágenes de las manos
# Aplicar operaciones lógicas
and_img = cv2.bitwise_and(img1, img2)
or_img = cv2.bitwise_or(img1, img2)
not_img1 = cv2.bitwise_not(img1)
not_img2 = cv2.bitwise_not(img2)  # Apply NOT to img2
xor_img = cv2.bitwise_xor(img1, img2)

# Mostrar las imágenes
titles = ["Imagen 1", "Imagen 2", "AND", "OR", "NOT (Imagen 1)", "NOT (Imagen 2)", "XOR"]
images = [img1, img2, and_img, or_img, not_img1, not_img2, xor_img]  # Include not_img2

plt.figure(figsize=(15, 6))  # Adjust figure size to accommodate more images

for i in range(len(images)):  # Loop through all images
    plt.subplot(2, 4, i + 1)  # Arrange in 2 rows, 4 columns
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```

***1. ¿Cómo puede ayudar la suma de imágenes en la combinación de datos de distintas modalidades médicas?***

La imagenología médica, por su parte, considera un conjunto de modalidades de
adquisición de imágenes médicas, las cuales se diferencian en cuanto a la naturaleza de los
principios físicos involucrados en el proceso de adquisición.
Las imágenes digitales representan información visual asociada con una escena
ambiental real que correspondería a lo que observamos con el sentido de la vista o bien
información no visible pero que puede ser medida utilizando sensores apropiados tales como
radiación infraroja, ultravioleta, rayos X ultrasonidos, etc.

***2. ¿En qué situaciones se podría usar la resta de imágenes en el análisis biomédico?***

La técnica de resta de imágenes se utiliza en el análisis biomédico para resaltar cambios o estructuras específicas eliminando información redundante. Es muy común en neuroimagen funcional (como fMRI o PET) para detectar actividad cerebral, en angiografía por sustracción digital para visualizar vasos sanguíneos eliminando huesos o tejidos, y en el seguimiento de tratamientos para comparar imágenes antes y después, identificando cambios sutiles.

***3. ¿Cómo afectan las operaciones lógicas a la segmentación de imágenes médicas?***

Las operaciones lógicas como AND, OR, NOT y XOR son fundamentales en la segmentación de imágenes médicas, ya que permiten combinar, refinar o comparar regiones de interés. La operación AND se usa para aislar zonas comunes entre dos máscaras (por ejemplo, tumores dentro del cerebro), OR para unir regiones segmentadas por diferentes métodos, NOT para excluir áreas no deseadas como el fondo, y XOR para resaltar diferencias entre segmentaciones (como validar una segmentación automática contra una manual).
